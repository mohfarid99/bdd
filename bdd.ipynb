```python
from ultralytics import YOLO

# Paths
data_path = r'D:\Naufal\acne\Blood Cells.v1i.yolov8\data.yaml'
```

```python
from ultralytics import YOLO

# Load and train the YOLOv8n model
model_n = YOLO('yolov8n.pt')  # Automatically downloads the model if not available
model_n.train(data=data_path, epochs=100, imgsz=640, batch=16, name='yolov8_n_bdd')
model_n.save('yolov8_n_bdd.pt')
```
